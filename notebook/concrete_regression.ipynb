### Model prep

```python
x = df.drop(['strength'], axis = 1).values
y = df['strength'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr_pred = lr.predict(x_test)
lr_pred[0:10]
```

```python
y_test[0:10]
```

```python
print("Training: ", round(lr.score(x_train, y_train)*100,2))
print("Testing: ", round(lr.score(x_test, y_test)*100,2))
```

```python
plt.figure(figsize = (12,4))
sns.scatterplot(lr_pred, color = 'r')
sns.scatterplot(y_test, color = 'g')
plt.show()
```

### Decision tree

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# Importing the packages
from sklearn.tree import DecisionTreeRegressor
decision_tree = DecisionTreeRegressor(max_depth = 7)

# Fitting the train model
decision_tree.fit(x_train, y_train)

# Predicting from the model 
DT_y_pred = decision_tree.predict(x_test)
DT_y_train = decision_tree.predict(x_train)

# Finding the error
MSE = mean_squared_error(y_test,DT_y_pred)
print("MSE:", MSE)

# Root mean squared error
RMSE = np.sqrt(MSE)
print("RMSE :" , RMSE)

# Adjusted r2
r2 = r2_score(y_test, DT_y_pred)
print("R2 for Decision Tree Regressor : ", r2*100, "\n")

print("Training: ", (decision_tree.score(x_train, y_train))*100)
print("Testing: ", (decision_tree.score(x_test, y_test))*100)
```

```python
plt.figure(figsize = (12,4))
sns.scatterplot(DT_y_pred, color = 'r')
sns.scatterplot(y_test, color = 'g')
plt.show()
```

### Random forest

```python
# Importing the model
from sklearn.ensemble import RandomForestRegressor

# Setting the hyperparameter values
random_forest = RandomForestRegressor(n_estimators = 25, max_depth=8)
random_forest.fit(x_train,y_train)
rf_y_pred = random_forest.predict(x_test)

# Mean squared error
MSE = mean_squared_error(y_test, rf_y_pred)
print("MSE :", MSE)

# Root mean squared error
RMSE = np.sqrt(MSE)
print("RMSE :", RMSE)

# Adjusted R2
r2 = r2_score(y_test, rf_y_pred)
print("R2 for Random Forest :", r2*100, "\n")


print("Training: ", (random_forest.score(x_train, y_train))*100)
print("Testing: ", (random_forest.score(x_test, y_test))*100)
```

```python
plt.figure(figsize = (12,4))
sns.scatterplot(rf_y_pred, color = 'r')
sns.scatterplot(y_test, color = 'g')
plt.show()
```

### Support Vector Regressor

```python
from sklearn.svm import SVR
svr = SVR()
```

```python
svr.fit(x_train, y_train)
svr_pred = svr.predict(x_test)
print(svr.score(x_train,y_train)*100,"%")
print(svr.score(x_test,y_test)*100,"%")
```

```python
plt.figure(figsize = (12,4))
sns.scatterplot(svr_pred, color = 'r')
sns.scatterplot(y_test, color = 'g')
plt.show()
```

### Gradient boosting

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
```

```python
GBR = GradientBoostingRegressor(learning_rate = 0.2, max_depth = 1, n_estimators = 800)
GBR.fit(x_train, y_train)
GBR_pred = GBR.predict(x_test)
print(GBR.score(x_train,y_train)*100,"%")
print(GBR.score(x_test,y_test)*100,"%")
```

```python
plt.figure(figsize = (12,4))
sns.scatterplot(GBR_pred, color = 'r')
sns.scatterplot(y_test, color = 'g')
plt.show()
```

### Overall score

```python
overall_scores = pd.DataFrame({'Linear Regression': ((lr.score(x_train, y_train))*100, (lr.score(x_test, y_test))*100),
                              'Decision Tree Regressor': ((decision_tree.score(x_train, y_train))*100, (decision_tree.score(x_test, y_test))*100),
                              'Random Forest Regressor': ((random_forest.score(x_train, y_train))*100, (random_forest.score(x_test, y_test))*100),
                              'Support Vector Regressor': ((svr.score(x_train, y_train))*100, (svr.score(x_test, y_test))*100),
                              'XGBoost Regressor': ((GBR.score(x_train, y_train))*100, (GBR.score(x_test, y_test))*100)})
overall_scores.T.rename(columns = {0:'Training Score', 1 : 'Test Score'})
```
